```python
raw_data = pd.read_csv('/Users/hanhyeseung/hshan/basic_work/0_data/data/arizona_pv_data.csv')
raw_data.set_index('dt', inplace=True)
raw_data.rename(columns={'Power(MW)':'pv_power'}, inplace=True)
```

## A. id를 column 이름으로 하는 dataframe 만들기

### 방법 1) for문으로 만들기

```python
pv_data = raw_data.copy()
```

```python
start_time = time.time()
pv_data_1 = pd.DataFrame()
for i in pv_data['plant_id'].unique():

    data = pv_data[pv_data['plant_id'] == i]
    pv_data_1['{}'.format(i)] = data['pv_power']
    
print('Elapsed Time: ', time.time()-start_time)
```

```python
pv_data_1.head()
```

### groupby로 만들기

```python
start_time = time.time()
pv_data_2 = pv_data.groupby('plant_id').apply(lambda x: x.pv_power).T

print('Elapsed Time: ', time.time()-start_time)
```
